```python
import skimage.color, skimage.io
from skimage.io import imread
from skimage.color import rgb2gray
from skimage import exposure, feature,img_as_ubyte
from skimage.filters.rank import entropy
from skimage.morphology import disk
from matplotlib import pyplot as plt 
import glob, os, warnings, cv2
import numpy as np
from scipy.ndimage.filters import median_filter
from numpy import array
import matplotlib.image as mpimg
from skimage.filters import threshold_otsu
import pandas as pd
from skimage.feature import local_binary_pattern
from scipy import misc
import matplotlib
```

```python
print("LBP after median (05,20)")

with open("0520_Acacia_lbpaftermed.csv", "w") as fo:

    fo.write("filename,data,class\n")

    for i in glob.glob("../../Image Dataset/Data Acacia/*"):
                
        print(i)

        label = i.split("-")[1].split(".")[0]
        label = 0 if label == 'S' else 1
        src_fname, ext = os.path.splitext(i)  

        #show original image
        ori = imread(i)
    
        #change to gray
        im = cv2.imread(i, cv2.IMREAD_GRAYSCALE)
        #save_fname_m = os.path.join(og0520, os.path.basename(src_fname)+'.JPG')
        #misc.imsave(save_fname_m, im)
        
        #contrast
        icon = exposure.rescale_intensity(im, in_range=(np.percentile(im,5), np.percentile(im,20)))
        with warnings.catch_warnings():
            warnings.simplefilter("ignore")
            dt = img_as_ubyte(icon)
        #save_fname_m = os.path.join(oc0520, os.path.basename(src_fname)+'.JPG')
        #misc.imsave(save_fname_m, icon)

        # make image blur
        imed = cv2.medianBlur(icon.astype(np.uint8), 3)
        #save_fname_m = os.path.join(om0520, os.path.basename(src_fname)+'.JPG')
        #misc.imsave(save_fname_m, imed)
        
        #LBP
        radius = 3
        n_points = 8 * radius
        lbp = local_binary_pattern(imed, n_points, radius, 'ror')
        #save_fname_m = os.path.join(olm0520, os.path.basename(src_fname)+'.JPG')
        #misc.imsave(save_fname_m, lbp)
        
        
        data = '-'.join(map(str,lbp.flatten().tolist()))
        
        fo.write("{},{},{}\n".format(i, data, label))
```

```python

print("lbp after thresholding")
with open("0520_Acacia_lbpafterthreshold.csv", "w") as fo:

    fo.write("filename,data,class\n")

    for i in glob.glob("../../Image Dataset/Data Acacia/*"):
                
        print(i)

        label = i.split("-")[1].split(".")[0]
        label = 0 if label == 'S' else 1
        src_fname, ext = os.path.splitext(i)  

        
        #show original image
        ori = imread(i)
    
        #change to gray
        im = cv2.imread(i, cv2.IMREAD_GRAYSCALE)
        
        #contrast
        icon = exposure.rescale_intensity(im, in_range=(np.percentile(im, 5), np.percentile(im,20)))
        with warnings.catch_warnings():
            warnings.simplefilter("ignore")
            dt = img_as_ubyte(icon)

        # make image blur
        imed = cv2.medianBlur(icon.astype(np.uint8), 3)
       
        #treshold
        threshold_val= threshold_otsu(imed)
        thresh = cv2.threshold(imed, threshold_val, 1, cv2.THRESH_BINARY)[1]
        #save_fname_m = os.path.join(ot0520, os.path.basename(src_fname)+'.JPG')
        #misc.imsave(save_fname_m, thresh)
        
        radius = 3
        n_points = 8 * radius
        lbp = local_binary_pattern(thresh, n_points, radius, 'ror')
        #save_fname_m = os.path.join(olt0520, os.path.basename(src_fname)+'.JPG')
        #misc.imsave(save_fname_m, lbp)
        
        data = '-'.join(map(str, lbp.flatten().tolist()))

        fo.write("{},{},{}\n".format(i, data, label))
```
